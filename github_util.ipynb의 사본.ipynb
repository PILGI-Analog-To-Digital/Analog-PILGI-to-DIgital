1. token 등록

```python
from google.colab import drive
from os.path import join  
## 여기다 token 두기
GIT_TOKEN = "" 
##
ROOT = "/content/gdrive/"
drive.mount(ROOT)

ORGANIZATION_NAME = "PILGI-Analog-To-Digital"
GIT_REPOSITORY = "Analog-PILGI-to-DIgital" 

GOOGLE_DRIVE_PATH = "/content/gdrive/Shareddrives/machine_learning_in_practice/Analog-PILGI-to-DIgital"

PROJECT_PATH = join(ROOT, GOOGLE_DRIVE_PATH)
print("PROJECT_PATH: ", PROJECT_PATH)   

GIT_PATH = "https://" + GIT_TOKEN + "@github.com/" + ORGANIZATION_NAME + "/" + GIT_REPOSITORY + ".git"
print("GIT_PATH: ", GIT_PATH)

%cd "{PROJECT_PATH}"
```
